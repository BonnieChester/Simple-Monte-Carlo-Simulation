This is my coursework project, involving the creation of a Monte Carlo model for profit forecasting under specified prices and costs. The pricing and material cost follow a normal distribution, while production cost adhere to a uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation times
simulations = 10000  # times

# The price and material cost
mean_price = 80
std_price = ( 90 - 70 ) / 6 # standard difference setting
mean_cost = 40
std_cost = ( 50 - 30 ) / 6  # standard difference setting

# production cost
min_production_cost = 18
max_production_cost = 22

# モンテカロシミュレーション
revenue_values = []
for _ in range(simulations):
    price = np.random.normal(mean_price, std_price)
    cost = np.random.normal(mean_cost, std_cost)
    production_cost = np.random.uniform(min_production_cost, max_production_cost)
    revenue = price - production_cost - cost
    revenue_values.append(revenue)

# plotting
plt.figure(figsize=(5, 2))
plt.hist(revenue_values, bins=100, color='blue', edgecolor='black', density = 'False')
plt.axhline(y = 0.02, color = 'black', linestyle = "--")
plt.axhline(y = 0.04, color = 'black', linestyle = "--")
plt.axhline(y = 0.06, color = 'black', linestyle = "--")
plt.title('Distribution of Product A')
plt.xlabel('Revenue')
plt.ylabel('Probability')

sorted_revenues = np.sort(revenue_values)
cumulative_probability = np.arange(1, simulations + 1) / simulations
plt.figure(figsize=(5, 2))
plt.plot(sorted_revenues, cumulative_probability, color='blue')
plt.title('Cumulative Probability Distribution of Prodcut A')
plt.axhline(y = 0.25, color = 'black', linestyle = "--")
plt.axhline(y = 0.5, color = 'black', linestyle = "--")
plt.axhline(y = 0.75, color = 'black', linestyle = "--")
plt.xlabel('Revenue')
plt.ylabel('Cumulative Probability')

plt.show()
```
